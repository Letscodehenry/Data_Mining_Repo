```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from scipy.stats import f_oneway
warnings.filterwarnings('ignore')
import folium
from folium.plugins import HeatMap
```

```python
df = pd.read_csv('dataset\\Global_terrorist.csv',encoding='ISO--8859-1')
```

<h1>Visualizing Geospatial with folium library</h1>

```python
# Filtering Data
filter1 = df[df['region_txt'] == 'Middle East & North Africa']
filter2 = filter1[filter1['iyear'] >= 2000]
filter3 = filter2[filter2['attacktype1_txt'] == 'Bombing/Explosion']
filter4 = filter3[filter3['nkill'] >= 10]
spatial = filter4.groupby(['longitude', 'latitude']).size().reset_index(name = 'Incident Count')

# Prepare Heat Data
heat_data = spatial[['latitude','longitude','Incident Count']].values.tolist()

# Create a base map
m = folium.Map(location=[30,40],zoom_start = 4)

# Add heatmap layer
HeatMap(heat_data).add_to(m)

# Save or Display the map
m.save('terrorism_heatmap.html')
m
```

<h1>Regional Analysis</h1>

```python
region_assin = df[(df['region_txt'] == 'Southeast Asia') & (df['attacktype1_txt'] == 'Assassination')]
year_kill = region_assin[(region_assin['iyear']>=2000) & (region_assin['nkill']>=5)] 

regional_spatial = year_kill.groupby(['latitude','longitude']).size().reset_index(name = 'Kill Count')

regional_heat_data = regional_spatial[['latitude','longitude','Kill Count']]

regional_base_map = folium.Map(location=[30,40],zoom_start = 4)

HeatMap(regional_heat_data).add_to(regional_base_map)

regional_base_map
```
