```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from scipy.stats import f_oneway
warnings.filterwarnings('ignore')
```

```python
dengue = pd.read_csv('dataset\\dengue.csv')
```

<h1>>>What month has the highest cases of dengue on year 2020?</h1>
<h1>>>What month has the lowest cases of dengue on year 2020?</h1>
<h1>>>How many cases of dengue occur on year 2020?</h1>

```python
year_2020 = dengue[dengue['Year']== 2020]
month_2020 = year_2020.groupby('Month')['Dengue_Cases'].sum()

month_2020.plot(kind='bar', color="Teal", edgecolor="black")
plt.gcf().set_size_inches(13,8)
plt.xlabel("Month")
plt.ylabel("Total Dengue Cases")
plt.title("Dengue Cases Per Month in 2020")
plt.xticks(rotation=45)
plt.show()

# To get the total Dengue Cases on Year 2020.
print(f"Dengue Cases on the Year 2020:'{month_2020.sum()}'")
```

```python
# To get total Dengue Cases of month
feb_2020 = year_2020[year_2020['Month'] == 'February']
cases_feb_2020 = feb_2020.groupby('Month')['Dengue_Cases'].sum()
cases_feb_2020
```

<h3><span style = color:Cyan >Insight No:3</span>  The total Dengue Cases on the Year 2020 is 91,041.</h3>

<h1>>>What are the top 5 region that has the highest cases of dengue?  </h1>

```python
every_region_cases = dengue.groupby('Region')['Dengue_Cases'].sum().reset_index()
top_5_highest_cases = every_region_cases.nlargest(5,'Dengue_Cases')

plt.barh(top_5_highest_cases['Region'], top_5_highest_cases['Dengue_Cases'],  color="Teal", edgecolor="black")
plt.title('Top 5 Region for Highest Cases of Dengue')
plt.xlabel('Dengue Cases')
plt.ylabel('Regions')
plt.gcf().set_size_inches(13,8)
plt.show()
```

```python
top_5_highest_cases
```

<h3><span style = color:Cyan >Insight No:10</span> The top 5 Regions that has the highest cases of dengue are Region IV-A that has '163,029', Region III that has '131,064', Region VI that has '117,523', NCR that has '115,966', and Region VII that has '110,683'. </h3>

<h1>>>What are the top 5 region that has the lowest cases of dengue?  </h1>

```python
top_5_lowest_cases = every_region_cases.nsmallest(5,'Dengue_Cases')

plt.barh(top_5_lowest_cases['Region'], top_5_lowest_cases['Dengue_Cases'],  color="Teal", edgecolor="black")
plt.title('Top 5 Region for Lowest Cases of Dengue')
plt.xlabel('Dengue Cases')
plt.ylabel('Regions')
plt.gcf().set_size_inches(13,8)
plt.show()
```

```python
top_5_lowest_cases
```
